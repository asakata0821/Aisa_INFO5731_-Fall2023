```python
import requests
from bs4 import BeautifulSoup
import time

information = []
narrator_interview_links = []
interview_download_links = []


for i in range(1,40):
  headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

  url = f'https://ddr.densho.org/narrators/?page={i}'
  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.content, 'html.parser')
  result = soup.find_all(class_='item-box')

  # Extract 'href' attributes from anchor elements within each item-box
  narrator_pages = [element.find('a')['href'] for element in result]

  for page_url in narrator_pages:
      headers = {
          'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
      narrator_response = requests.get(page_url, headers=headers)
      narrator_soup = BeautifulSoup(narrator_response.content, 'html.parser')

      # Print the HTML content of the page for analysis
      print(f'Collecting data from {page_url}')


      # Extract biodata
      narrator_names = narrator_soup.find('h1').text.replace('\n','')

      narrator_introductions = narrator_soup.find('p').text


      b_tags = narrator_soup.find_all('b')
      for b_tag in b_tags:
          a_tag_interview = b_tag.find('a')
          if a_tag_interview and 'href' in a_tag_interview.attrs:
              narrator_interview_links.append(a_tag_interview['href'])

      for interview_page_url in narrator_interview_links:
        headers = {
          'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
        interview_response = requests.get(interview_page_url, headers=headers)
        interview_soup = BeautifulSoup(interview_response.content, 'html.parser')

        a_tags = interview_soup.find_all('a', class_='btn btn-default btn-xs')
        for a_tag_download in a_tags:
              # Check if 'href' attribute exists before accessing it

            interview_download_link = a_tag_download['href']
            if interview_download_link.startswith('https://archive.org/download/ddr-densho') and interview_download_link.endswith('.mpg'):
                      interview_download_links.append(interview_download_link)





      information.append([narrator_names, narrator_introductions])


      # Add a delay between requests
      time.sleep(5)
```

```python
import pandas as pd
df = pd.DataFrame(information, columns = ['Name', 'Introduction'])
df
```
