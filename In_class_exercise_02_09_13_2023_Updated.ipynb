Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#This is for the first page only
book_list = []

for i in range(1,51):
  book_url = f'https://books.toscrape.com/catalogue/page-{i}.html'
  book_response = requests.get(book_url)
  book_response = book_response.content
  book_soup = BeautifulSoup(book_response, 'html.parser')
  book_results = book_soup.find('ol')

  books = book_results.find_all('article', class_='product_pod')

  for book in books:
    image = book.find('img')
    title = image.attrs['alt']
    star = book.find('p')
    star = star['class'][1]
    price = book.find('p', class_='price_color').text
    book_list.append([title, price, star])
```

```python
book_df = pd.DataFrame(book_list, columns = ['Title', 'Price', 'Rating (out of 5 stars)'])
book_df
#If we want to save, then we can run df.to_csv('Book.csv')
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
#This is for the first page only

firstpage_url = 'https://dl.acm.org/action/doSearch?AllField=information+retrieval&AfterYear=2013&BeforeYear=2023&startPage=&ContentItemType=research-article&pageSize=20'
response = requests.get(firstpage_url)
response = response.content
soup = BeautifulSoup(response, 'html')
results = soup.find_all(class_ = 'search__item issue-item-container')

titles = []
authors = []
years = []
conferences = []
links = []


for item in results:
    title = item.select('h5')[0].get_text()
    titles.append(title)

    authors = [a['title'] for a in soup.select('ul.rlist--inline.loa a')]
    authors.append(authors)

    year = item.select('.dot-separator')[0].get_text() #year, later make a list and append, print[1]
    years.append(year)

    conference = item.select('.epub-section__title')[0].get_text()
    conferences.append(conference)

link_parts = soup.find_all(class_ = 'hlFld-Title')
for link_line in link_parts:
  link_line = link_line.find_all('a')
  for link in link_line:
    link = link.get('href')
    links.append('https://dl.acm.org' + link)



print(titles)
print(authors)
print(conferences)
print(links)
```

```python
#This is for the second page to the 50th page (till the 1000th data)

for i in range(1,5):
  url = f'https://dl.acm.org/action/doSearch?AllField=information+retrieval&AfterYear=2013&BeforeYear=2023&startPage={i}&ContentItemType=research-article&pageSize=20'
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html')
  results = soup.find_all(class_ = 'search__item issue-item-container')

  titles = []
  authors = []
  years = []
  conferences = []
  links = []


  for item in results:
      title = item.select('h5')[0].get_text()
      titles.append(title)

      authors = [a['title'] for a in soup.select('ul.rlist--inline.loa a')]
      authors.append(authors)

      year = item.select('.dot-separator')[0].get_text() #year, later make a list and append, print[1]
      years.append(year)

      conference = item.select('.epub-section__title')[0].get_text()
      conferences.append(conference)

  link_parts = soup.find_all(class_ = 'hlFld-Title')
  for link_line in link_parts:
    link_line = link_line.find_all('a')
    for link in link_line:
      link = link.get('href')
      links.append('https://dl.acm.org' + link)
  i += 1


  print(titles)
  print(authors)
  print(conferences)
  print(links)
  print('------------------------------------------')
```

```python
#This is the collection of abstracts

abstracts = []

for page in links:
  response2 = requests.get(page)
  response2 = response2.content
  soup2 = BeautifulSoup(response2,'html.parser')

  abstract_results = soup2.find_all(class_ = 'abstractSection abstractInFull')

  for abstract in abstract_results:
    paragraphs = abstract.find_all('p')
    for paragraph in paragraphs:
      abstract = paragraph.get_text()
      abstracts.append(abstract)

print(abstracts)



# Untill here, question (3)
```
