```python
#This one
import os
import zipfile
import pandas as pd

def collect_transcripts_by_category(df, category_0, category_1, category_2, category_3, category_4, category_5, category_6, category_7):
    labels = df['predicted_labels'].dropna()

    for index, row in df.iterrows():
        transcript = row['Transcript (Narrator)']
        label = row['predicted_labels']

        if label == 0:
            category_0.append(transcript)
        elif label == 1:
            category_1.append(transcript)
        elif label == 2:
            category_2.append(transcript)
        elif label == 3:
            category_3.append(transcript)
        elif label == 4:
            category_4.append(transcript)
        elif label == 5:
            category_5.append(transcript)
        elif label == 6:
            category_6.append(transcript)
        elif label == 7:
            category_7.append(transcript)

category_0 = []
category_1 = []
category_2 = []
category_3 = []
category_4 = []
category_5 = []
category_6 = []
category_7 = []

#for i in range(4, 5):  # For Densho
zip_file_name = 'Category_Labeled_JASC(Interviewer).zip'
with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    for filename in zip_file.namelist():
        if filename.endswith('.csv'):
            with zip_file.open(filename) as csv_file:
                df = pd.read_csv(csv_file, encoding='utf-8')

                collect_transcripts_by_category(
                    df,
                    category_0,
                    category_1,
                    category_2,
                    category_3,
                    category_4,
                    category_5,
                    category_6,
                    category_7
                )
```

```python
df_category_0 = pd.DataFrame(category_0)
df_category_1 = pd.DataFrame(category_1)
df_category_2 = pd.DataFrame(category_2)
df_category_3 = pd.DataFrame(category_3)
df_category_4 = pd.DataFrame(category_4)
df_category_5 = pd.DataFrame(category_5)
df_category_6 = pd.DataFrame(category_6)
df_category_7 = pd.DataFrame(category_7)

df_category_0.to_csv('JASC_category_0.csv', index=False)
df_category_1.to_csv('JASC_category_1.csv', index=False)
df_category_2.to_csv('JASC_category_2.csv', index=False)
df_category_3.to_csv('JASC_category_3.csv', index=False)
df_category_4.to_csv('JASC_category_4.csv', index=False)
df_category_5.to_csv('JASC_category_5.csv', index=False)
df_category_6.to_csv('JASC_category_6.csv', index=False)
df_category_7.to_csv('JASC_category_7.csv', index=False)
```
